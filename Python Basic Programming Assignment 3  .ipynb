```python
#Ans--4
w = int(input("Enter number is check if its prime number "))
if w > 1 :
    for i in range(2,w):
        if w%i == 0:
            print("Number", w  ,"is not prime number")
            break
    else:
            print("Number", w ," is prime number")
else:
    print("Number is not prime number")
```
